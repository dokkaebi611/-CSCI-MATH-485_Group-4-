### Check dataset

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('loan_approval_dataset.csv')
df.head()
```

```python
print(df.columns)
print(df.shape)
```

```python
## Check whether there are any missing values in the dataset
df.isnull().sum()
```

```python
info = df.describe()
info
```

```python
info = info.drop(index='count', errors='ignore')
info.plot(kind='box', subplots=True, layout=(4,4), figsize=(12,12))

## we can see that there are some outliers in the dataset
```

```python
# Check the distribution of the target variable with proportion labels
status_prop = df[' loan_status'].value_counts(normalize=True)

ax = status_prop.plot(kind='bar')

for i, value in enumerate(status_prop.values):
    ax.text(i, value + 0.02 , f'{value:.1%}')

plt.show()

## we say that the dataset is slightly inbalanced
```
